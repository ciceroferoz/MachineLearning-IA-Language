```python
# Version 1.0
# This is a simple algoritmo of Python using IA to learning whatever Language that you want. 
# This document use the most frequency words used in a conversation.
# No mind what topic is and the most frequency words are 80% of the conversation
# You can read whatever text once you learnd these words



# Bibliotecas necessrias
import nltk
import pandas as pd
from nltk.probability import FreqDist
from string import punctuation
%matplotlib inline
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

!curl https://raw.githubusercontent.com/ciceroferoz/MachineLearning-IA-Language/master/podcast.txt -o podcast.txt
doc1 = open("podcast.txt", "r")

# Faz leitura do documento e print o contedo
docTxt = doc1.read()
print(docTxt)
# Remove numeros no texto
txt = ''.join(n for n in docTxt if not n.isdigit())
# remove os pontos no texto e coloca as letra em maiusculas
txt = ''.join(n for n in txt if n not in punctuation).lower()
# Imprimir o texto j tratado
print (txt)
nltk.download("punkt")
# Separa o texto em palavras individuais
palavra = nltk.tokenize.word_tokenize(txt)
# Reune as palavras mais frequentes em um data frame
fdist = FreqDist(palavra)
count_frame = pd.DataFrame(fdist, index =[0]).T
count_frame.columns = ['Count']
print (count_frame)

counts = count_frame.sort_values('Count', ascending = False)
fig = plt.figure(figsize=(16, 9))
ax = fig.gca()    
counts['Count'][:80].plot(kind = 'bar', ax = ax, color="blue")
ax.set_title('Palavras mais frequentes no texto')
ax.set_ylabel('Palavras frequentes')
ax.set_xlabel('Palavras')
plt.show()


# remove palavras que no relevantes
nltk.download("stopwords")
from nltk.corpus import stopwords
txt = ' '.join([word for word in txt.split() if word not in (stopwords.words('english'))])
print("\n")
print(txt)




print("\n")
print(txt)
# Reune as palavras mais frequentes restantes no texto
palavra = nltk.tokenize.word_tokenize(txt)
fdist = FreqDist(palavra)
count_frame = pd.DataFrame(fdist, index =[0]).T
count_frame.columns = ['Count']
# Mostra as 80 palavras mais importantes
counts = count_frame.sort_values('Count', ascending = False)
fig = plt.figure(figsize=(16, 9))
ax = fig.gca()    
counts['Count'][:80].plot(kind = 'bar', ax = ax, color="blue")
ax.set_title('80 Palavras mais usadas no texto')
ax.set_ylabel('Palavras frequentes')
ax.set_xlabel('Palavras')
plt.show()

 

```
